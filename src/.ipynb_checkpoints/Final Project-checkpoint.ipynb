```python
I = IP.rgb2gray(IP.imread("Doyley.jpeg"))
IP.imshow(I, cmap = "gray")
I.shape
```

```python
############# making subimages ################

i = 0
j = 0
rows = int(644/161)
columns = int(512/64)
array = []
start_x = start_y = 0
end_x = 161
end_y = 64

### using code from homework 1 ###
### the loop iterates over (48,101) each time starting from (0,0) ###
### each iteration of (48,101,36) is one subimage which is cropped and stored in array ###

for i in range(rows):
    start_y = 0
    end_y = 64
    for j in range(columns):
        array.append(I[start_x:end_x, start_y:end_y])
        start_y = start_y + 64;
        end_y = end_y + 64;
    start_x = start_x + 161;
    end_x = end_x + 161;
    
fig = IP.plt.figure(figsize = (20,20))
d = 1
e=0
for i in range(rows):
    for j in range(columns):
        axis = fig.add_subplot(rows,columns,d)
        axis.imshow(array[e], cmap = "gray", aspect = "auto")
        d = d+1
        e = e+1

```

```python
############# applying otsu thresholding to subimages ##############

array_otsuthreshold = []  ## array to store subimages after applying Otsu Thresholding
jj = 0
while jj < len(array):
    thres_otsu = IP.greythres(array[jj])
    array_otsuthreshold.append(array[jj] > thres_otsu)
    jj = jj + 1
```

```python
############# displaying subimages after otsu thresholding ##############



fig = IP.plt.figure(figsize = (20,20))
d = 1
e=0
for i in range(rows):
    for j in range(columns):
        axis = fig.add_subplot(rows,columns,d)
        axis.imshow(array_otsuthreshold[e], cmap = "gray", aspect = "auto")
        d = d+1
        e = e+1
```

```python
############ recombining sub-images ##################
### using code from homework 1 ###

row_array = [0] * 8
temp = array_otsuthreshold[0]
c = j = 0
for i in range(0,rows):
    temp = array_otsuthreshold[c]
    c = c + 1
    for j in range(columns - 1):
        temp = IP.np.hstack((temp, array_otsuthreshold[c]))
        c = c + 1
    row_array[i] = temp

comb_img = row_array[0]
for i in range(1, 4):
    comb_img = IP.np.vstack((comb_img, row_array[i]))

IP.plt.figure(figsize=(10,10)); 
IP.imshow(comb_img, cmap = 'gray'); IP.plt.title("Recombination after Otsu Thresholding")
```
